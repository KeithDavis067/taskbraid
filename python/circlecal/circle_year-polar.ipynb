```python
y = Year_Data(2024)
ydict = y.to_dict()
```

```python
fig = go.Figure()
d_to_degree = (360 / len(y))
dbase=np.array([0.95] * y.length().days)
dr=1 - dbase
fig.add_trace(go.Barpolar(base=dbase,
                          r=dr,
                          theta0 = -90,
                          dtheta = d_to_degree,
                          name="date",
                          hovertext=[d.date() for d in ydict["date"]],
                          #text = [d.day for d in ydict["date"]]
                         )
             )
#mbase=np.array([0.95] * y.length().days)
#mr=1 - base
#fig.add_trace(go.Barpolar(
#                          theta0 = -90,
#                          dtheta = (360 / len(month_name[1:])),
#                          name="month",
#                          hovertext=[d.date() for d in ydict["date"]]))
fig.update_layout(height=1000)
fig.show()
```
