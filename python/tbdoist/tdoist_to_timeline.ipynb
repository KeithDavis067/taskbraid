```python
tasks  = api.get_tasks()
```

```python
df = pd.DataFrame(tasks)
```

```python
df.loc[69]["due"]
```

```python
def catch_none_factory(key):
    def catch_none(obj):
        try:
            return obj[key]
        except TypeError:
            return None
    return catch_none
    
for key in ["date", "is_recurring", "string", "datetime", "timezone"]:
    catch_none = catch_none_factory(key)
    df[key] = df["due"].apply(catch_none)
```
